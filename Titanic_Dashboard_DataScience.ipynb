```python
from jupyter_dash import JupyterDash 
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
```

```python
import pandas as pd
import numpy as np
import plotly
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
#reading dataset
```

```python
df=pd.read_csv('/content/titanic.csv')
df
```

```python
#Some Preprocessing of Data
```

```python
cols=['PassengerId','Name','Ticket','Cabin']
df=df.drop(cols,axis=1)
df=df.dropna()
df['family']=df.iloc[:,4]+df.iloc[:,5]
df=df.drop(['SibSp','Parch'],axis=1)
df
```

```python
#making graphs via Plotly
```

```python
fig1 = go.Figure(data=[go.Pie(labels=df['Survived'],pull=0.1,title='Survival Rate')]) 
fig1.update_layout(title="Survival Rate")

print("figure 1 done")
```

```python
fig2 = go.Figure(data=[go.Pie(labels=df['Sex'], values=df['Survived'],hole=0.3)])
fig2.update_layout(title="Genders Per Survival")  

fig3 = make_subplots(rows=1, cols=2, specs=[[{"type": "pie"}, {"type": "pie"}]])
fig3.add_trace( go.Pie(labels=df.loc[df['Sex'] == 'male']['Survived'], pull = [.1, .1],
                   title = 'Males Survived'), row=1, col=1)

fig3.add_trace(
            go.Pie(labels=df.loc[df['Sex'] == 'female']['Survived'], pull = [.07, .07],
                   title = 'Females Survived'),row=1, col=2)

fig3.update_layout(title="Survival Per Genders") 

fig5 = px.violin(df, x='Pclass', y="Age", color='Survived')
fig5.update_layout(title='Distribution of passengers distinguished by survived per age for each Pclass')

fig6 = px.scatter(df, x="Age", y="Fare",color="Survived" ,title='Age Vs Fare depicted differentiated on Survival',color_continuous_midpoint=1)

fig7 = px.line(df, np.arange(0,712), y="Fare",color="Pclass", title='Fare variation on Different Pclass')
```

```python
# making Dash app
```

```python
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H1("Titanic Data Set"),
    html.H2("Harneet Kaur"),
    html.H2("101917189"),
    html.H2("CSE-8"),
   # html.H3("Survival Pie Chart"),
    dcc.Graph(id='graph',figure=fig1),

        html.P("Column:"),
    dcc.Dropdown(
        id='col', 
        value='Age', 
        options=[{'value': x, 'label': x} 
                 for x in ['Age', 'Fare', 'Sex', 'Pclass','family','Embarked']],
        clearable=False
    ),

    dcc.Graph(id='graphd1'),

    dcc.Graph(id='graph2',figure=fig2), 

    dcc.Graph(id='graph3',figure=fig3), 

    dcc.Graph(id='graph5',figure=fig5), 

    dcc.Graph(id='graph6',figure=fig6), 
    
    dcc.Graph(id='graph7',figure=fig7), 


])
@app.callback(
    Output('graphd1', 'figure'), 
    Input('col','value'))
def generate_chart(col):
      fig = px.histogram(df, col, color='Survived')
      return fig

```

```python
app.run_server()
```
